```python
import numpy as np
import pandas as pd
import folium
import branca

from folium.plugins import HeatMap
```

```python
df1 = pd.read_excel('./data/export_content_x_uk_x_filter_nativeretweets_geocode_55_378051_3_43597299999999_750km_2016_04_21_11_46_43.xlsx',
                    sheetname='Top Hashtags')
```

```python
hashtags = df1[['Hash']].copy()
```

```python
df2 = pd.read_excel('./data/export_dashboard_x_uk_x_filter_nativeretweets_geocode_55_378051_3_43597299999999_750km_2016_04_21_10_32_03.xlsx',
                    sheetname='Stream')
```

```python
df2['Latitude'] = df2['Latitude'].astype(float)
df2['Longitude'] = df2['Longitude'].astype(float)
```

```python
twits = df2[['Tweet content', 'Latitude', 'Longitude']].copy()
```

```python
hashtags
```

```python
hashname = "#nowplaying"
```

```python
content_with_hash = twits[twits['Tweet content'].str.contains(hashname)]
```

```python
heat_data = content_with_hash[['Latitude', 'Longitude']].values.tolist()
```

```python
map_hash = folium.Map(location=[51.30, 0.7], zoom_start=2)
```

```python
HeatMap(heat_data).add_to(map_hash)
```

```python
map_hash
```

```python
f = branca.element.Figure()

for i in range(1):
    name = "#" + hashtags['Hash'].iloc[i]
    twitswhash = twits[twits['Tweet content'].str.contains(name)]
    heat_data = twitswhash[['Latitude', 'Longitude']].values.tolist()
    
    map_tweet = folium.Map(location=[51.30, 0.7], zoom_start=2)
    HeatMap(heat_data).add_to(map_tweet)
    
    d = f.add_subplot(4, 1, i+1)
    d.add_child(map_tweet)
```
