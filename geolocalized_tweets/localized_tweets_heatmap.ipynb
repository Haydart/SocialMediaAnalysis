```python
import branca
import folium
import pandas as pd
from folium.plugins import HeatMap
```

```python
top_hashtags_df = pd.read_excel(
    './export_content_x_uk_x_filter_nativeretweets_geocode_55_378051_3_43597299999999_750km_2016_04_21_11_46_43.xlsx',
    sheet_name='Top Hashtags'
)
tweets_df = pd.read_excel(
    './export_dashboard_x_uk_x_filter_nativeretweets_geocode_55_378051_3_43597299999999_750km_2016_04_21_10_32_03.xlsx',
    sheet_name='Stream'
)

tweets_df['Latitude'] = tweets_df['Latitude'].astype(float)
tweets_df['Longitude'] = tweets_df['Longitude'].astype(float)

print(top_hashtags_df.head())
print(tweets_df.head())
```

```python
hashtags_df = top_hashtags_df[['Hash']].copy()
print(hashtags_df.head())
```

```python
tweets_location_df = tweets_df[['Tweet content', 'Latitude', 'Longitude']].copy()
hashname = "#nowplaying" # most popular, just for testing the single heatmap
```

```python
tweets_containing_hashes_df = tweets_location_df[tweets_location_df['Tweet content'].str.contains(hashname)]
print(tweets_containing_hashes_df.head())
```

```python
heatmap_data = tweets_containing_hashes_df[['Latitude', 'Longitude']].values.tolist()
folium_map = folium.Map(location=[53.5, 0], zoom_start=5)
HeatMap(heatmap_data).add_to(folium_map)
folium_map
```

```python
figure = branca.element.Figure()

for i in range(4):
    name = "#" + hashtags_df['Hash'].iloc[i]
    tweets_hash_df = tweets_location_df[tweets_location_df['Tweet content'].str.contains(name)]
    heatmap_data = tweets_hash_df[['Latitude', 'Longitude']].values.tolist()
    
    tweet_map = folium.Map(location=[53.5, 0], zoom_start=3)
    HeatMap(heatmap_data).add_to(tweet_map)
    
    d = figure.add_subplot(4, 1, i + 1)
    d.add_child(tweet_map)

figure
```
