```python
%matplotlib inline
import community
import matplotlib.pyplot as plt
import networkx as nx

G = nx.erdos_renyi_graph(100, 0.05)

#compute the best partition
partition = community.best_partition(G)

#drawing
size = float(len(set(partition.values())))
pos = nx.spring_layout(G)
count = 0.
partition_set = set(partition.values())

for community in partition_set:
    count = count + 1.
    nodes_list = [nodes for nodes in partition.keys() if partition[nodes] == community]
    colors_list = range(len(nodes_list))
    nx.draw_networkx_nodes(G, pos, nodes_list, node_size=20, node_color=colors_list)
nx.draw_networkx_edges(G, pos, alpha=0.5)
plt.show()
```
